```python
df_empate = df[['temporada', 'empate']]
df_empate = df_empate.loc[(df_empate['empate']==1.0)]
df_empate['empate'].count()
```

```python
df_casa = df[['temporada', 'vitoria_casa']]
df_casa = df_casa.loc[(df_casa['vitoria_casa']==1.0)]
df_casa['vitoria_casa'].count()
```

```python
df_fora = df[['temporada', 'vitoria_fora']]
df_fora = df_fora.loc[(df_fora['vitoria_fora']==1.0)]
df_fora['vitoria_fora'].count()
```

```python
df_fora.head()
```

```python
plt.style.use('seaborn')
```

```python
resultado = ['casa', 'fora', 'empate']
lista = [2,3,4]

x = np.array(lista)

def autolabel(grupos):
    for i in grupos:
        h = i.get_height()
        ax.annotate('{}'.format(h),
                    xy = (i.get_x() + i.get_width()/2, h),
                    xytext = (0,3),
                    textcoords = 'offset points',
                    ha = 'center')

fig,ax = plt.subplots(figsize=(5,3))

largura = 0.50

grupo1 = ax.bar(df_fora['vitoria_fora'] + 2, df_fora['vitoria_fora'].value_counts(), largura, label='Fora', color='r')
grupo2 = ax.bar(df_casa['vitoria_casa'] + 1, df_casa['vitoria_casa'].value_counts(), largura, label='Casa', color='b')
grupo3 = ax.bar(df_empate['empate'] + 3, df_empate['empate'].value_counts(), largura, label='Empate', color='g')

ax.legend()
ax.set_title('VITORIA TIME CASA, VITORIA TIME FORA E EMPATE ENTRE 2005 A 2022', fontsize=10)
ax.set_ylabel('QUANTIDADE')

ax.set_ylim([0, 2700])
ax.set_xticks(x)
ax.set_xticklabels(resultado)

autolabel(grupo1)
autolabel(grupo2)
autolabel(grupo3)


plt.show()
```

```python
df_temporada_casa = df_casa.groupby('temporada')
df_temporada_casa.sum()
```

```python
df_temporada_fora = df_fora.groupby('temporada')
df_temporada_fora.sum()
```

```python
df_temporada_empate = df_empate.groupby('temporada')
df_temporada_empate.sum()
```

```python
dfcasa = df_temporada_casa.sum()
dffora = df_temporada_fora.sum()
dfempate = df_temporada_empate.sum()
```

```python
temporadas = [2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]

fig,ax = plt.subplots(figsize=(12,5))

largura = 0.30

grupocasa = ax.bar(dfcasa.index + largura*-1, dfcasa['vitoria_casa'], largura, label='casa', color='b')
grupofora = ax.bar(dffora.index + largura*0, dffora['vitoria_fora'], largura, label='fora', color='r')
grupoempate = ax.bar(dfempate.index + largura*1, dfempate['empate'], largura, label='empate', color='g')

ax.legend()
ax.set_title('VITORIAS TIME CASA, VITORIAS TIME FORA E EMPATES POR TEMPORADA')
ax.set_ylabel('QUANTIDADE')
ax.set_ylim(0,200)

ax.set_xticks(temporadas)

autolabel(grupocasa)
autolabel(grupofora)
autolabel(grupoempate)

plt.savefig('graficos.png')
plt.show()
```

```python
dftimecasa = df[['time_casa', 'gols_casa', 'gols_fora', 'vitoria_casa','vitoria_fora', 'empate']]

dftimecasa.columns = ['time_casa', 'gols feitos', 'gols sofridos', 'vitorias', 'derrotas', 'empates']

dftimecasa = dftimecasa.groupby('time_casa').sum().sort_values(by='gols feitos', ascending=True)

dftimecasa['jogos'] = dftimecasa['vitorias'] + dftimecasa['derrotas'] + dftimecasa['empates']
dftimecasa['media gols feitos'] = dftimecasa['gols feitos'] / dftimecasa['jogos']
dftimecasa['media gols sofridos'] = dftimecasa['gols sofridos'] / dftimecasa['jogos']

display(dftimecasa)
```

```python
dftimefora = df[['time_fora', 'gols_fora', 'gols_casa', 'vitoria_fora', 'vitoria_casa', 'empate']]

dftimefora.columns = ['time_fora', 'gols feitos', 'gols sofridos', 'vitorias', 'derrotas', 'empates']

dftimefora = dftimefora.groupby(['time_fora']).sum().sort_values(by='gols feitos', ascending=True)

dftimefora['jogos'] = dftimefora['vitorias'] + dftimefora['derrotas'] + dftimefora['empates']
dftimefora['media gols feitos'] = dftimefora['gols feitos'] / dftimefora['jogos']
dftimefora['media gols sofridos'] = dftimefora['gols sofridos'] / dftimefora['jogos']

display(dftimefora)
```

```python
dfgolstemporada = df[['temporada', 'gols_casa', 'gols_fora', 'vitoria_casa', 'vitoria_fora', 'empate']]
dfgolstemporada = dfgolstemporada.groupby(['temporada']).sum()

dfgolstemporada['gols'] = dfgolstemporada['gols_casa'] + dfgolstemporada['gols_fora']
dfgolstemporada['jogos'] = dfgolstemporada['vitoria_casa'] + dfgolstemporada['vitoria_fora'] + dfgolstemporada['empate']
dfgolstemporada['media gols'] = dfgolstemporada['gols'] / dfgolstemporada['jogos']

dfgolstemporada
```
